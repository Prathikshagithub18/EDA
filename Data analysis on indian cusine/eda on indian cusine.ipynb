### Total number of vegetarian and non-vegetarian dishes

```python
ID['diet'].value_counts().plot(kind='bar', figsize=(8,5))
plt.xlabel("Diet", fontsize=15)
plt.ylabel("total number of vegetarian", fontsize=15)
plt.show()
```

From above graph  we can clearly say that vegetarian are more than non vegetarain.

### About Non-vegetarian items

```python
nv=ID[['name','state','cook_time','region']]
nv_fet=ID.diet=='non vegetarian'
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data=nv1_df, x = "name",y='cook_time')
plt.xticks(rotation=90)
plt.xlabel("name", fontsize=15)
plt.ylabel("cook_time",fontsize=15)
plt.show()
```

### Gujarat Snacks

```python
sk=g1_df[['name','cook_time','prep_time','course']]
sk_df=g1_df.course=='snack'
```

```python
sk1=sk[sk_df]
sk1_df=sk1.sort_values(by=['cook_time'],ascending=False)
sk1_df
```

```python
na=sk1_df.name
ctime=sk1_df.cook_time
ptime=sk1_df.prep_time
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=80)
plt.title('Gujarat Snacks')
plt.plot(na,ctime,marker='*',mfc='r',mec='y',markersize=10,linewidth=4,linestyle='--')
plt.plot(na,ptime,marker='o',mfc='r',mec='y',markersize=10,linewidth=4,linestyle='--')
plt.ylabel('cook_time & prep_time' )
plt.legend(['cook_time','prep_time'],fontsize=18)

```

The above plot shows the comparsion of snacks between cook_time and prep_time dishes that are present in the state Gujarat

### Gujarat main course

```python
mc=g1_df[['name','cook_time','prep_time','course']]
mc_df=g1_df.course=='main course'
```

```python
mc1=mc[mc_df]
mc1_df=mc1.sort_values(by=['cook_time'],ascending=False)
mc1_df
```

```python
na1=mc1_df.name
ctime=mc1_df.cook_time
ptime=mc1_df.prep_time
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=60)
plt.title('Gujarat Main course')
plt.plot(na1,ctime,marker='*',mfc='r',mec='y',markersize=10,linewidth=4,linestyle='--',color='purple')
plt.plot(na1,ptime,marker='o',mfc='r',mec='y',markersize=10,linewidth=4,linestyle='--',color='aqua')
plt.ylabel('cook_time & prep_time' )
plt.legend(['cook_time','prep_time'],fontsize=18)
```

The above plot shows the comparsion of main-course between cook_time and prep_time dishes that are present in the state Gujarat.

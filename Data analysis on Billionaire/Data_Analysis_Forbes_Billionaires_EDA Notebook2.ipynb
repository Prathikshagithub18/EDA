### <div style='color:rgb(255, 255, 255);background-color:rgba(10, 5, 5, 0.85);padding:10px 10px;;font-size:25px;line-height:2'>Which Source has the highest NetWorth ?</div>

### <div style='color:rgb(255, 255, 255);background-color:rgba(10, 5, 5, 0.85);padding:10px 10px;;font-size:25px;line-height:2'>Which Person has the highest NetWorth ?</div>

```python
plt.figure(figsize=(15,8))
sns.barplot(data=df, x = "Networth", y = "Name", order = df['Name'].head(20), estimator = sum , errwidth = 0)
plt.xticks(rotation=90)
plt.xlabel("Networth", fontsize=15)
plt.ylabel("Name", fontsize=15)
plt.show()
```

### <div style='color:rgb(255, 255, 255);background-color:rgba(10, 5, 5, 0.85);padding:10px 10px;;font-size:25px;line-height:2'>Top 5 Billionaires from China ?</div>

```python
chinese = df[df['Country'] == 'China']
top_5_chinese = chinese.sort_values('Networth',ascending=False).head()# declaring exploding pie
explode = [0.1, 0.01, 0, 0, 0]
palette_color = sns.color_palette('dark')
plt.pie(top_5_chinese['Networth'], labels=top_5_chinese['Name'], colors=palette_color,
        explode=explode, autopct='%.0f%%')
plt.show()
```
